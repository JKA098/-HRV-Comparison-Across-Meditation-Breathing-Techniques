# Combine chi_med and chi_pre together

```python

file_C1_med = r"C:\Users\johan\OneDrive\Documents\Sharing file\School_Grad\0.2025 courses\winter\AMOD_5310H_Artificial Intelligence\Project\Project\dataset\o_0_chi\C1_meditation.csv"
file_C1_pre = r"C:\Users\johan\OneDrive\Documents\Sharing file\School_Grad\0.2025 courses\winter\AMOD_5310H_Artificial Intelligence\Project\Project\dataset\o_0_chi\C1_pre_meditation.csv"

# Read both datasets
df_med = pd.read_csv(file_C1_med)
df_pre = pd.read_csv(file_C1_pre)

# Ensure both DataFrames have the same structure
expected_columns = ["Time", "HeartRate", "Subject", "Group", "State"]
df_med = df_med[expected_columns]
df_pre = df_pre[expected_columns]


# Merge the two DataFrames[4].
c1_merged_df = pd.concat([df_pre, df_med], ignore_index=True)

# Define output file path
merged_file = r"C:\Users\johan\OneDrive\Documents\Sharing file\School_Grad\0.2025 courses\winter\AMOD_5310H_Artificial Intelligence\Project\Project\dataset\o_0_chi\C1_merged_data.csv"


# Save merged dataset as CSV
c1_merged_df.to_csv(merged_file, index=False)

print(f"✅ Merged CSV file saved successfully at {merged_file}!")
```

```python
df_c1_merged_df = pd.DataFrame(c1_merged_df)
#df_c1_merged_df.head()
```

### basic plot example

```python
import matplotlib.pyplot as plt
```

```python
# Convert "Time" 
df_c1_merged_df["Time"] = df_c1_merged_df["Time"].astype(float)

# Filter for only Subject C1 and C1_pre
df_C1 = df_c1_merged_df[df_c1_merged_df["Subject"] == "C1"]
df_C1_pre = df_c1_merged_df[df_c1_merged_df["Subject"] == "C1_pre"]

# Separate Meditation and Pre-Meditation
df_meditation = df_c1_merged_df[df_c1_merged_df["State"] == "Meditation"]
df_pre_meditation = df_c1_merged_df[df_c1_merged_df["State"] == "Pre_Meditation"]

# Create the plot
plt.figure(figsize=(12, 6))

# Plot Meditation data
plt.plot(df_meditation["Time"], df_meditation["HeartRate"], label="Meditation", color="blue", alpha=0.7)

# Plot Pre-Meditation data
plt.plot(df_pre_meditation["Time"], df_pre_meditation["HeartRate"], label="Pre-Meditation", color="red", alpha=0.7)

# Labels and title
plt.xlabel("Time")
plt.ylabel("Heart Rate (BPM)")
plt.title("Heart Rate - Meditation vs. Pre-Meditation for Chi group")
plt.legend()

# Show the plot
plt.show()
```
