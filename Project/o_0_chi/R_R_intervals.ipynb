```python

# Function to load QRS peak annotations
def load_qrs_peaks(qrs_file):
    
    annotation = wfdb.rdann(qrs_file, 'qrs')  # Read QRS annotations
    r_peaks = annotation.sample  # Extract QRS peak locations
    return r_peaks

# Function to compute R-R intervals
def compute_rr_intervals(r_peaks, sampling_rate=250):
    
    rr_intervals = np.diff(r_peaks) / sampling_rate  # Convert to seconds
    return rr_intervals

# Example usage
qrs_file = "heart-rate-oscillations-during-meditation-1.0.0\data\C1med"  # Replace with actual path (without .qrs extension)


# Load QRS peaks
r_peaks = load_qrs_peaks(qrs_file)

# Compute R-R intervals
rr_intervals = compute_rr_intervals(r_peaks)

# Display first few values
print("First 10 R-R Intervals (seconds):", rr_intervals[:10])

# Plot R-R intervals over time
plt.figure(figsize=(10, 5))
plt.plot(rr_intervals, marker='o', linestyle='-')
plt.xlabel("Beat Number")
plt.ylabel("R-R Interval (s)")
plt.title("R-R Intervals from QRS Peaks")
plt.grid()
plt.show()
```
